### 📊 SYSTEM ENTITY RELATIONSHIP DIAGRAM (ERD)

| Entity Name | Primary Key (PK) | Foreign Keys (FK) | Core Attributes | Relationships |
| :--- | :--- | :--- | :--- | :--- |
| **Users** | `UserID` | *None* | Username, Email | One-to-Many with *Applicant_Details* |
| **Applicant_Details** | `ApplicantID` | `UserID` | Income, Age | One-to-One with *Credit_History*, One-to-Many with *Predictions* |
| **Credit_History** | `HistoryID` | `ApplicantID` | Credit_Score, Prior_Default | Linked to *Applicant_Details* |
| **ML_Model** | `ModelID` | *None* | Model_Name, Version | One-to-Many with *Predictions* |
| **Approval_Prediction** | `PredictionID` | `ApplicantID`, `ModelID` | Approval_Status | Connects Applicants to the ML Model |
