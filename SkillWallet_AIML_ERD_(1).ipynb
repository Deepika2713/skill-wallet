# SkillWallet - AIML Domain ER Diagram
Generates the ER diagram (Users, User_Profile, Skill_Assessment, ML_Model, Skill_Prediction) using Graphviz.

```python
!pip install graphviz -q
```

```python
from graphviz import Digraph

dot = Digraph('SkillWallet_AIML_ERD', format='png')
dot.attr(rankdir='TB', splines='ortho', fontsize='11', fontname='Helvetica')
dot.attr('node', shape='box', fontname='Helvetica', fontsize='11', margin='0.25,0.15')

dot.node('Users', 'Users\n(UserID, Name, Email, Password, Role)')
dot.node('User_Profile', 'User_Profile\n(ProfileID, UserID, SkillCategory,\nExperienceLevel, EducationType,\nEmploymentStatus, ActiveDays)')
dot.node('ML_Model', 'ML_Model\n(ModelID, ModelName, Version,\nTrainingDate, Accuracy)')
dot.node('Skill_Assessment', 'Skill_Assessment\n(AssessmentID, ProfileID,\nAssessmentType, Score, CompletedDate)')
dot.node('Skill_Prediction', 'Skill_Prediction\n(PredictionID, ProfileID, ModelID,\nPredictedSkillLevel, ConfidenceScore,\nPredictionDate)')

dot.edge('Users', 'User_Profile', label='1        N')
dot.edge('User_Profile', 'Skill_Assessment', label='1        N')
dot.edge('User_Profile', 'Skill_Prediction', label='1        1')
dot.edge('ML_Model', 'Skill_Prediction', label='1        N')

dot.render('skillwallet_aiml_erd', view=False)
dot
```
